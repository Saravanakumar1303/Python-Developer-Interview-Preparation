Given a number n, find the first digit of the number.

Input: n = 123
Output: 1

```python
def firstDigit(n):
    n = abs(n)
    while n>=10:
        n//=10
    return n

n = int(input())
result = firstDigit(n)
print(result)
```

Given an positive integer n, print numbers from 1 to n without using loops.

Implement the function printTillN() to print the numbers from 1 to n as space-separated integers.

Input: n = 5
Output: 1 2 3 4 5
Explanation: We have to print numbers from 1 to 5.

```python
class Solution:
    def printTillN(self,n):
        if n == 0:
            return
        self.printTillN(n-1)
        print(n, end=" ")

n = int(input())
Solution().printTillN(n)
```

First n Fibonacci using DP

You are given an integer n, return the fibonacci series till the nth(0-based indexing) term. Since the terms can become very large return the terms modulo 10**9+7.

Input: n = 5
Output: [0, 1, 1, 2, 3, 5]

```python
class Solution:
    def fibbonacciNumbers(self,n):
        mod = 10**9+7
        fib = [0] * (n+1)
        if n>=1:
            fib[1] = 1
        for i in range(2,n+1):
            fib[i] = (fib[i-1]+fib[i-2]) % mod
        return fib

n = int(input())
sol = Solution()
result = sol.fibbonacciNumbers(n)
print(result)
```

Juggler Sequence is a series of integers in which the first term starts with a positive integer number a and the remaining terms are generated from the immediate previous term using the below recurrence relation:
Odd number -> sqrt(n) *3
even number -> sqrt(n)

Input: n = 9
Output: 9 27 140 11 36 6 2 1

Input: n = 6
Output: 6 2 1
Explaination: 
[61/2] = 2. 
[21/2] = 1.
Constraints:
1 ≤ n ≤ 100

Expected Complexities

```python
import math
class Solution:
    def jugglerSequence(self, n):
        seq = [n]
        while n!=1:
            if n % 2 == 0:
                n = math.isqrt(n)
            else:
                n = math.isqrt(n**3)
            seq.append(n)
        return seq

n = int(input())
sol = Solution()
output = sol.jugglerSequence(n)
print(output)
```

Given a number n, find the value of n raised to the power of its own reverse.
Note: The result will always fit into a 32-bit signed integer.

Input: n = 10
Output: 10
Explanation: The reverse of 10 is 1 (leading zero is discarded), and 10 raised to the power 1 is 10.

```python
class Solution:
    def reverseExponentiation(self, n):
        reverse = int(str(n)[::-1])
        return n ** reverse
    
n = int(input())
sol = Solution()
output = sol.reverseExponentiation(n)
print(output)
```
