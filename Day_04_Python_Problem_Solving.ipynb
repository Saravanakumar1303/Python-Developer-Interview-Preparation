Repeat the Strings - Python

Given two strings a and b. The task is to make a new string where the string with longer length should be in between and the one with shorter length should be outside on front and end. New string should be like shorter+longer+shorter.

Input: 
a = Hi, b = There
Output: 
HiThereHi
Explanation: 
After joining short+long+short strings, 
we have new string as "HiThereHi".

```python
def combo_string(a, b):
    if len(a) <= len(b):
       short = a
       longer = b
    else:
        short = b
        longer = a
    return short + longer + short

a,b = input().split()
output = combo_string(a,b)
print(output)
```

String Functions I - Python

In this question, we'll take a look at inbuilt string methods like title(), swapcase(), find(), strip().
You'll be given a string str and x, you'll perform various operations on them.

Input:
str = hello 
x = llo
Output:
hello 
2 
Hello 
HELLO

```python
def trim(str):
    print(str.strip())


def exists(str, x):
    print(str.find(x))


def titleIt(str):
    print(str.title())


def casesSwap(str):
    print(str.swapcase())

str,x = input().split()
trim(str)
exists(str,x)
titleIt(str)
casesSwap(str)
```

Count Vowels and consonants

```python
str = input("Enter the string").replace(" ","")
print(str)
vowels_list =('a','e','i','o','u','A','E','I','O','U')
vowels_count = 0
consonants_count = 0
vowels = []
consonants = []
for s in str:
    if s in vowels_list:
        vowels.append(s)
        vowels_count +=1
    else:
        consonants.append(s)
        consonants_count +=1
print(f"vowels_count:{vowels_count} consonants_counts :{consonants_count}")
print(f"vowels : {vowels}, consonants : {consonants}")
```

Check Anagram

```python
s1 = input("Enter the first string:")
s2 = input("Enter the second string:")
if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")


from collections import Counter

print(Counter(s1))
print(Counter(s2))
```

```python
words = ["eat","tea","tan","ate","nat","bat"]

from collections import defaultdict

anagrams = defaultdict(list)
print(anagrams)

for word in words:
    key = "".join(sorted(word))
    #print("key",key)
    anagrams[key].append(word)
print(anagrams.values())
```

First Non repeating Character

```python
from collections import Counter
str = input()
count = Counter(str)

for ch in str:
    if count[ch] == 1:
        print(ch)
        break
```
